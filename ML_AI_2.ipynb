```python
import pandas as pd
import warnings
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
df0 = pd.read_csv('titanic.csv')
```

```python
df0.head()
```

```python
df = df0.drop(columns = ["Name"])
```

```python
df
```

```python
warnings.filterwarnings("ignore")

list_of_features_string = ['Sex']

transformer = make_column_transformer(
    (OneHotEncoder(), list_of_features_string),
    remainder='passthrough')

transformer_one_hot = transformer.fit_transform(df)
transformer_one_hot_df = pd.DataFrame(transformer_one_hot, columns=transformer.get_feature_names())
```

```python
transformer_one_hot_df
```

```python
X_new = transformer_one_hot_df.drop(columns = ["Survived"])
y_new = transformer_one_hot_df["Survived"]
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.8)
```

طبقه‌بندی

```python
classifer = GaussianNB()
classifer.fit(X_train_new, y_train_new)
y_predict_new = classifer.predict(X_test_new)
```

ارزیابی

```python
accuracy = accuracy_score(y_test_new, y_predict_new)
```

```python
accuracy
```

```python
precision = precision_score(y_test_new, y_predict_new, average='macro')
```

```python
precision
```
