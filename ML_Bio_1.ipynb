انتخاب برچسب‌ها و تقسیم داده به بخش‌های آموزش و آزمون

```python
import sklearn 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = ["Survived"])
y = df["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)
```

طبقه‌بندی

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifer = GaussianNB()
classifer.fit(X_train, y_train)
y_predict = classifer.predict(X_test)
```

را‌ه‌حل ارور:
    پیش پردازش

حذف ستون‌های اضافی

```python
df_Droped_Name = df.drop(columns = ["Name"])
```

```python
df_Droped_Name
```

تغییر مقدارهای غیر عددی به عددی

```python
df_Droped_Name_Int_coding = df_Droped_Name
```

```python
df_Droped_Name_Int_coding
```

```python
df_Droped_Name_Int_coding["Sex"].replace('female',0, inplace=True)
df_Droped_Name_Int_coding["Sex"].replace('male',1, inplace=True)
```

```python
df_Droped_Name_Int_coding
```

مجددا انتخاب برچسب‌ها و تقسیم داده به بخش‌های آموزش و آزمون

```python
X_new = df_Droped_Name_Int_coding.drop(columns = ["Survived"])
y_new = df_Droped_Name_Int_coding["Survived"]
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.8)
```

طبقه‌بندی

```python
classifer = GaussianNB()
classifer.fit(X_train_new, y_train_new)
y_predict_new = classifer.predict(X_test_new)
```

ارزیابی

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
accuracy = accuracy_score(y_test_new, y_predict_new)
```

```python
accuracy
```

```python
precision = precision_score(y_test_new, y_predict_new, average='macro')
```

```python
precision
```
